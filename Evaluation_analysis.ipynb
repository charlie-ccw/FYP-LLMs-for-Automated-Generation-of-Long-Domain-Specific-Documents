```python
import os
import json
import matplotlib.pyplot as plt
import pandas as pd
```

```python
version_1_results = {}
version_1_path = "/content/evaluation_result/version_1"
files = os.listdir(version_1_path)

for file in files:
  file_path = os.path.join(version_1_path, file)
  base_name = os.path.splitext(file)[0]
  with open(file_path, 'r', encoding='utf-8') as f:
    data = json.load(f)

  version_1_results[base_name] = data
```

```python
version_2_results = {}
version_2_path = "/content/evaluation_result/version_2"
files = os.listdir(version_2_path)

for file in files:
  file_path = os.path.join(version_2_path, file)
  base_name = os.path.splitext(file)[0]
  with open(file_path, 'r', encoding='utf-8') as f:
    data = json.load(f)

  version_2_results[base_name] = data
```

```python
version_3_results = {}
version_3_path = "/content/evaluation_result/version_3"
files = os.listdir(version_3_path)

for file in files:
  file_path = os.path.join(version_3_path, file)
  base_name = os.path.splitext(file)[0]
  with open(file_path, 'r', encoding='utf-8') as f:
    data = json.load(f)

  version_3_results[base_name] = data
```

```python
version_3_1_results = {}
version_3_1_path = "/content/evaluation_result/version_3_1"
files = os.listdir(version_3_1_path)

for file in files:
  file_path = os.path.join(version_3_1_path, file)
  base_name = os.path.splitext(file)[0]
  with open(file_path, 'r', encoding='utf-8') as f:
    data = json.load(f)

  version_3_1_results[base_name] = data
```

```python
section_ids = ['1.1', '1.2', '1.3', '1.4', '1.5', '1.6', '1.7', '1.8', '1.9', '1.10', '1.11', '1.12', '1.13', '1.14', '1.15', '1.16', '1.17', '1.18', '2.1', '2.2', '2.3', '2.4', '2.5', '3.1', '3.2', '3.3', '3.4', '3.5', '3.6', '4.1', '4.2', '4.3', '4.4', '5.1', '5.2', '5.3', 'final']
```

# BLEU

```python
bleu_results = {}
for id in section_ids:
  bleu_results[id] = {
      'version_1': [],
      'version_2': [],
      'version_3': [],
      'version_3_1': []
  }
for file, file_results in version_1_results.items():
  for section_id, score in file_results['BLEU'].items():
    bleu_results[section_id]['version_1'].append(score)

for file, file_results in version_2_results.items():
  for section_id, score in file_results['BLEU'].items():
    bleu_results[section_id]['version_2'].append(score)

for file, file_results in version_3_results.items():
  for section_id, score in file_results['BLEU'].items():
    bleu_results[section_id]['version_3'].append(score)

for file, file_results in version_3_1_results.items():
  for section_id, score in file_results['BLEU'].items():
    bleu_results[section_id]['version_3_1'].append(score)
```

```python
for section_id, datas in bleu_results.items():
  data = [datas['version_1'], datas['version_3_1'], datas['version_2'], datas['version_3']]
  plt.boxplot(data, patch_artist=True, whis=100)
  plt.title(f'BLEU Box Plot of {section_id}')
  plt.xlabel('Version')
  plt.ylabel('Values')
  plt.xticks([1, 2, 3, 4], ['version 1', 'version 3_1', 'version 2', 'version 3'])

  plt.show()
```

# GPT_SELF_EVALUATION

```python
gpt_self_evaluation_results = {}
for id in section_ids:
  gpt_self_evaluation_results[id] = {
      'version_1': [],
      'version_2': [],
      'version_3': [],
      'version_3_1': []
  }
for file, file_results in version_1_results.items():
  for section_id, result in file_results['GPT_SELF_EVALUATION'].items():
    score = result['score'] if section_id != 'final' else result
    gpt_self_evaluation_results[section_id]['version_1'].append(score)

for file, file_results in version_2_results.items():
  for section_id, result in file_results['GPT_SELF_EVALUATION'].items():
    score = result['score'] if section_id != 'final' else result
    gpt_self_evaluation_results[section_id]['version_2'].append(score)

for file, file_results in version_3_results.items():
  for section_id, result in file_results['GPT_SELF_EVALUATION'].items():
    score = result['score'] if section_id != 'final' else result
    gpt_self_evaluation_results[section_id]['version_3'].append(score)

for file, file_results in version_3_1_results.items():
  for section_id, result in file_results['GPT_SELF_EVALUATION'].items():
    score = result['score'] if section_id != 'final' else result
    gpt_self_evaluation_results[section_id]['version_3_1'].append(score)
```

```python
for section_id, datas in gpt_self_evaluation_results.items():
  data = [datas['version_1'], datas['version_3_1'], datas['version_2'], datas['version_3']]
  plt.boxplot(data, patch_artist=True, whis=100)
  plt.title(f'Rouge L Box Plot of {section_id}')
  plt.xlabel('Version')
  plt.ylabel('Values')
  plt.xticks([1, 2, 3, 4], ['version 1', 'version 3_1', 'version 2', 'version 3'])

  plt.show()
```

# Final All

```python
final_all_results = [bleu_results, rouge_l_results, meteor_results, bleurt_results, gpt_self_evaluation_results]
matrixs = ["BLEU", "ROUGE_L", "METEOR", "BLEURT", "GPT_SELF_EVALUATION"]
for i, results in enumerate(final_all_results):
  data = [results['final']['version_1'], results['final']['version_3_1'], results['final']['version_2'], results['final']['version_3']]
  plt.boxplot(data, patch_artist=True, whis=100)
  plt.title(f'{matrixs[i]} Box Plot of Final')
  plt.xlabel('Version')
  plt.ylabel('Values')
  plt.xticks([1, 2, 3, 4], ['version 1', 'version 3_1', 'version 2', 'version 3'])

  plt.show()
```

```python
final_all_results = [bleu_results, rouge_l_results, meteor_results, bleurt_results, gpt_self_evaluation_results]
matrixs = ["BLEU", "ROUGE_L", "METEOR", "BLEURT", "GPT_SELF_EVALUATION"]
evaluation_table = {}
for i, results in enumerate(final_all_results):
  data = [
      sum(results['final']['version_1'])/len(results['final']['version_1']),
      sum(results['final']['version_3_1'])/len(results['final']['version_3_1']),
      sum(results['final']['version_2'])/len(results['final']['version_2']),
      sum(results['final']['version_3'])/len(results['final']['version_3'])
      ]

  evaluation_table[matrixs[i]] = data

df = pd.DataFrame(evaluation_table, index=['version_1', 'version_3_1', 'version_2', 'version_3'])
df
```
